This jupyter notebook gives a touch of data sciency aproach to see what even is the 1_million_chess_games dataset

```python
import pandas
dataset_path = './games_cleaned.csv'
```

```python
df = pandas.read_csv(dataset_path)
df.head()
```

```python
# remove rows with NaN in 'Moves' column and keep only rows with games lasting more than one move
df = df[df['Moves'].notna() & (df['Moves'].str.len() > 6)]
```

```python
df.head()
```

```python
moves = df['Moves'].tolist()
print(len(moves))
print(type(moves[0]))
print(moves[0])
```

```python
# split the moves by inter and dot after
import re
parsed_moves = []
for move in moves:
    parsed_moves.append([])
    moves_splitted = re.findall(r'(\d+\. )(\S+ )(\S+)', move)
    for number, white_move, black_move in moves_splitted:
        parsed_moves[-1].append((int(number.replace(". ", "")), white_move.replace(" ", ""), black_move.replace(" ", "")))
print(parsed_moves[0:2])
```

```python
# average, min and max number of moves
num_moves = [len(move) for move in parsed_moves]
print(num_moves.index(min(num_moves)))
print(f'Average number of moves: {sum(num_moves) / len(num_moves)}')
print(f'Min number of moves: {min(num_moves)}')
print(f'Max number of moves: {max(num_moves)}')
```
