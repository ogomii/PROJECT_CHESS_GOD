```python
import pandas
import matplotlib.pyplot as plt
dataset_path = './train.csv'
```

```python
df = pandas.read_csv(dataset_path, sep=',')
print(df.describe())
print(df.info())
```

Architecture idea: </br>
each field represented by 12 classes (12 binary inputs as there is 6 white and 6 black pieces) </br>
input matrix (64, 12) </br>
each input matrix (12,) will be processed by a subnet into an embedded space of some dim d resulting in (64,d) </br>
(64,d), as it has a small outer dimention can easily be processed by a transformer network :)  </br>
